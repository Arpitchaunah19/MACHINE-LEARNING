```python
import re
import pandas as pd

# Function to split contact number and email from mixed string
def separate_contact_email(value):
    # Extract email using regex
    email_match = re.search(r'[\w\.-]+@[\w\.-]+', str(value))
    email = email_match.group(0) if email_match else ''
    
    # Remove email to get contact number
    contact = str(value).replace(email, '').strip()
    return pd.Series([contact, email])

# Apply function to the column and create NEW COLUMNS in DataFrame
patients_df[['Contact_Number', 'Email_Address']] = patients_df['email'].apply(separate_contact_email)

# Optionally drop the original mixed column
# patients_df.drop(columns='contact', inplace=True)

# Preview the new columns
print(patients_df[['Contact_Number', 'Email_Address']].head())
```
